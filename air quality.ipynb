```python
data1=df['Unique ID']   #finding outlier by iris
data1
q1 = np.percentile(data1,25)
print(q1)
```

```python
q3 = np.percentile(data1,75)
print(q3)
```

```python
iqr = q3-q1
print(iqr)
```

```python
ub=q3+(1.5*iqr)
print(ub)
lb = q1=(1.5*iqr)
print(lb)
outliers = data1[(data1 < lb) | (data1 > ub)]
print(outliers)
```

```python
#finding outlier by zscore
data2 = df['Unique ID']
data2
mean=np.mean(data2)
print("Mean Of Data",mean)
std=np.std(data2)
print("Print STD Of Data",std)
z_score=(data2-mean)/std
print(z_score)
print(z_score.head(16))
outliers=data2[(z_score < -3) | (z_score > 3)]
print("Printing Outliers",outliers)
print(data2.head(16))
```

```python
df = df.drop(outliers.index)  #drop rows containing outliers
outliers=data2[(z_score < -3) | (z_score > 3)]
print("Printing Outliers",outliers)
print(data2.head(16))
```
